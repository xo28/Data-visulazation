```python
import pandas as pd
data = pd.read_csv("GlobalLandTemperaturesByState.csv")
dictionary = {}

newlist=[]
x = len(data.index)

for x in range(0, x):
    if data["Country"][x] == "United States":
        newlist.append(data.loc[x])
states = pd.DataFrame(newlist)
states.reset_index(drop=True, inplace=True)

states.head(-10)
```

```python
year = 1902
state_list = []
date_list = []
temp_list = []

# Past: 1902-1922
for y in range(0, len(states)):
    if (int(states["dt"][y][0:4]) == year) and year<=1922:
        if year == 1922:
            for x in range(0,12):
                date_list.append(states["dt"][y+x])
                state_list.append(states["State"][y+x])
                temp_list.append(states["AverageTemperature"][y+x])
            year= 1902
        else:
            for x in range(0,12):
                date_list.append(states["dt"][y+x])
                state_list.append(states["State"][y+x])
                temp_list.append(states["AverageTemperature"][y+x])
            year +=1

year = 2002
# Present: 2002-2012
for t in range(0, len(states)):
    if (int(states["dt"][t][0:4]) == year) and year<=2012:
        if year == 2012:
            for k in range(0,12):
                date_list.append(states["dt"][t+k])
                state_list.append(states["State"][t+k])
                temp_list.append(states["AverageTemperature"][t+k])
            year = 2002
        else:
            for k in range(0,12):
                date_list.append(states["dt"][t+k])
                state_list.append(states["State"][t+k])
                temp_list.append(states["AverageTemperature"][t+k])
            year +=1
print(date_list)
```

```python
dictionary['Date'] = date_list
dictionary['Temperature'] = temp_list
dictionary['State'] = state_list
final = pd.DataFrame.from_dict(dictionary)
final.to_csv('testing.csv', header='true')
```
